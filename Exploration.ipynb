```python
#Data Exploration time first, what do I have here?
def get_csv():
    translations = pd.DataFrame(columns=['English', 'French'])
    with open('eng-fren.txt', 'r') as text:
        for line in text.readlines():
            translation = line.strip('\n').split('\t')
            phrase = {'English': translation[0], 'French':translation[1]}
            translations = translations.append(phrase, ignore_index=True)
    translations.to_csv('eng-fren.csv')
```
